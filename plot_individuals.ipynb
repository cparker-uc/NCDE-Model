```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from get_data import NelsonData, AblesonData
```

```python
nelson_control_dataset = NelsonData(patient_groups=['Control'], normalize_standardize='None')
nelson_atypical_dataset = NelsonData(patient_groups=['Atypical'], normalize_standardize='None')
nelson_melancholic_dataset = NelsonData(patient_groups=['Melancholic'], normalize_standardize='None')
nelson_neither_dataset = NelsonData(patient_groups=['Neither'], normalize_standardize='None')
```

```python
ableson_control_dataset = AblesonData(patient_groups=['Control'], normalize_standardize='None')
ableson_mdd_dataset = AblesonData(patient_groups=['MDD'], normalize_standardize='None')
```

## These are the Nelson Control patients that were classified correctly 100% of the time

```python
%matplotlib widget
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(10,10))
fig.canvas.toolbar_visible = True
for idx, (patient, _) in enumerate(nelson_control_dataset):
    if idx+1 in (1,6,7,8):
        ax1.plot(patient[:,0], patient[:,1], label=f'Patient {idx+1}')
        ax2.plot(patient[:,0], patient[:,2], label=f'Patient {idx+1}')
ax1.legend(shadow=True, fancybox=True, loc='upper right')
ax2.legend(shadow=True, fancybox=True, loc='upper right')
```
